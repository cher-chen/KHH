```python
import pandas as pd
from datetime import datetime, timedelta
pd.set_option('display.max_columns', 100)

df = pd.read_excel("data/khh1005.xlsx")
#
df_add = df.dropna(subset = ['tug2_no'])
df_add2 = df.dropna(subset = ['tug3_no'])

df_add['tug1_no'] = df_add['tug2_no']
df_add2['tug1_no'] = df_add2['tug3_no']

df_add = df_add.drop(columns = ['tug2_no','tug3_no'])
df_add2 = df_add2.drop(columns = ['tug2_no','tug3_no'])
df = df.drop(columns = ['tug2_no','tug3_no'])
df = pd.concat([df,df_add,df_add2],ignore_index=True)
df = df.sort_values(by='start_time')
cols = df.columns.tolist()
cols = ['tug1_no',
 'ship_no',
 'sailing_status',
 'min_work_time',
 'max_work_time',
 'mean_work_time',
 'start_time',
 'min_end_time',
 'max_end_time',
 'pilot_ready_time',
 'pilot_wait_time',
 'port',
 'place1',
 'place2',
 'pilot1',
 'pilot2',
 'tug_cnt',
 'total_weight',
 'front_weight',
 'back_weight',
 'weight_level',
 'pier_info',
 'dist',
 'wind',
 'park',
 'reverse',
 'seven',
 'day',
 'avg_hp']
df = df[cols]
df.rename(columns={"tug1_no" : "tug"}, inplace=True)
df.dropna(subset = ['tug','place2'])
df = df.reset_index()
```

```python
size = len(df)
tug_from = pd.Series([None] * size)
dic = {101:-1,
 104:-1,
 106:-1,
 108:-1,
 109:-1,
 112:-1,
 143.0:-1,
 145.0:-1,
 151.0:-1,
 152.0:-1,
 153.0:-1,
 155.0:-1,
 162.0:-1,
 163.0:-1,
 165.0:-1,
 171.0:-1,
 172.0:-1,
 181.0:-1,
 241.0:-1,
 245.0:-1,
 303.0:-1,
 306.0:-1,
 308.0:-1,
 321.0:-1,
 322.0:-1,
 401.0:-1,
 451.0:-1,
 568.0:-1}
df_tug = df.tug.get_values()
last = pd.Series([None] * size)

def metch_move_time():
    i = 0
    while i < size:
        if dic[df_tug[i]] == -1:
            dic[df_tug[i]] = i
            last[i] = -1
        elif dic[df_tug[i]] != -1:
            last[i] = df.place2[dic[df_tug[i]]]
            dic[df_tug[i]] = i
            
        i+=1
```

```python
last.columns = "last"
ndf = pd.concat((df,last.rename('last')),axis = 1)
ndf.dropna(subset = ['last'])
```

```python
dist_pier = pd.read_excel('data/complete_dis_meter.xlsx', index_col=1)
dist_pier
def fetch_dist(row):
    if row['sailing_status'] == 'I':
        if pd.isna(row['last']) or int(row['last']) not in dist_pier.index:
            return None
        return dist_pier.loc[int(row['last']), 9001] if int(row['port']) == 1 else dist_pier.loc[int(row['last']), 9002]
    
    elif row['sailing_status'] == 'O':
        if pd.isna(row['last']) and pd.isna(row['place2']) or int(row['last']) not in dist_pier.index or int(row['place2']) not in dist_pier.index:
            return None
        return dist_pier.loc[int(row['last']), int(row['place2'])]
    
    elif row['sailing_status'] == 'T':
        if pd.isna(row['place1']) or pd.isna(row['last']) or int(row['place1']) not in dist_pier.index or int(row['last']) not in dist_pier.index:
            return None
        return dist_pier.loc[int(row['last']), int(row['place1'])]

    
ndf['move_dist'] = ndf.apply(fetch_dist, axis=1)
# ndf = ndf.dropna(subset = ['move_dist'])
ndf['move_min'] = (ndf['move_dist'] / 18520) * 60
ndf
writer = pd.ExcelWriter('data/main1010.xlsx')
ndf.to_excel(writer,'Sheet1')
writer.save()
```
