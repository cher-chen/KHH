```python
import pandas as pd
data = pd.read_excel('data/main1010.xlsx')
```

```python
%matplotlib inline
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
I = data[data.sailing_status == 'I']
# I = I.dropna(subset = ["max_work_time"])
fig, ax = plt.subplots()
fig.set_size_inches(15, 10)

sns.distplot(I[(I.port == 1) & (I.weight_level <= 4)].mean_work_time.values, 
             kde=False, 
             hist_kws={"histtype": "step", "alpha": 1, "color": "red"},
             bins=range(1, 130, 5),
             label="A",
             ax=ax,
            )
sns.distplot(I[(I.port == 1) & (I.weight_level > 4)].mean_work_time.values, 
             kde=False, 
             hist_kws={"histtype": "step", "alpha": 1, "color": "orange"},
             bins=range(1, 130, 5),
             label="B",
            )
sns.distplot(I[(I.port == 2) & (I.weight_level <= 4)].mean_work_time.values, 
             kde=False, 
             hist_kws={"histtype": "step", "alpha": 1, "color": "green"},
             bins=range(1, 130, 5),
             label="C",
            )
plt.xlim(0, 130)
sns.distplot(I[(I.port == 2) & (I.weight_level > 4)].mean_work_time.values, 
             kde=False, 
             hist_kws={"histtype": "step", "alpha": 1, "color": "blue"},
             bins=range(1, 130, 5),
             label="D",
            )
plt.xlim(0, 130)
fig.savefig('I1all.png', dpi=100)
```
