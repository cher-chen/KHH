```python
import pandas as pd
data = pd.read_excel('data/main1010.xlsx')
%matplotlib inline
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
I = data[data.sailing_status == 'I']
fig, ax = plt.subplots()
fig.set_size_inches(15, 10)
ax1 = plt.subplot(221)
ax2 = plt.subplot(222)
ax3 = plt.subplot(223)
ax4 = plt.subplot(224)
r=sns.distplot(I[(I.port == 1) & (I.weight_level <= 4)].move_min, 
             kde=False, 
             hist_kws={ "alpha": 0.6, "color": "red"},
             bins=range(0, 40, 4),
             label="A",
             ax=ax1,
            )
ax1.legend(["port1/small"],fontsize=20)
r.set(xticks=(range(0,40,4)),xlim=(0,40),xlabel="work time(min)",ylabel="frequency",yticks=(range(0,3000,500)),ylim=(0,3000))

o=sns.distplot(I[(I.port == 1) & (I.weight_level > 4)].move_min, 
             kde=False, 
             hist_kws={"alpha": 0.6, "color": "orange"},
             bins=range(0, 40, 4),
             label="B",
             ax=ax2
            )
ax2.legend(["port1/large"],fontsize=20)
o.set(xticks=(range(0,40,4)),xlim=(0,40),xlabel="work time(min)",ylabel="frequency",yticks=(range(0,3000,500)),ylim=(0,3000))


g=sns.distplot(I[(I.port == 2) & (I.weight_level <= 4)].move_min, 
             kde=False, 
             hist_kws={"alpha": 0.6, "color": "green"},
             bins=range(0, 40, 4),
             label="C",
             ax=ax3
            )
ax3.legend(["port2/small"],fontsize=20)
g.set(xticks=(range(0,40,4)),xlim=(0,40),xlabel="work time(min)",ylabel="frequency",yticks=(range(0,3000,500)),ylim=(0,3000))

plt.xticks(range(0,130,4))
b=sns.distplot(I[(I.port == 2) & (I.weight_level > 4)].move_min, 
             kde=False, 
             hist_kws={ "alpha": 0.6, "color": "blue"},
             bins=range(0, 40, 4),
             label="D",
             ax=ax4
            )
ax4.legend(["port2/large"],fontsize=20)
b.set(xticks=(range(0,40,4)),xlim=(0,40),xlabel="work time(min)",ylabel="frequency",yticks=(range(0,3000,500)),ylim=(0,3000))

fig.savefig('mvIcomp.png', dpi=100)
```
